# PCA Start

```python
from sklearn               import decomposition
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

# (x- mean) / (sd)
# (x- xmin) / (xmax - xmin)
# Note -- before doing the PCA always make the data unitless

```python
scaler = StandardScaler()
```

```python
scaler_pa = scaler.fit_transform(pa)
```

```python
scaler_pa
# here we can see we have done the standardization as it has made the data unit less
```

# know we can start actual # PCA work

```python
pca_pa = PCA()
```

```python
data_af_PCA_trans = pca_pa.fit_transform(scaler_pa)
```
